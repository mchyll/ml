```python
%matplotlib inline

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
img = x_train[0]

print(img[5:6, 10:20])
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')

img = img / 255
print('After normalizing:')
print(img[5:6, 10:20])
plt.subplot(1, 2, 2)
plt.imshow(img, cmap='gray')

plt.show()
```
